## Rename NLSY79 Variables 

NLSY79 variables are downloaded with columns labeled by reference number; update to reflect question name (which includes survey year).

* Use DictReader to create a dataframe that thinks of each row of values as a dictionary, an individual list, with the first row (column name) as a key associated with the values. 

```python
# Import necessary packages
import csv
from csv import DictReader
from typing import List, Dict
import numpy as np
import pandas as pd 
```

```python
# Create a file handle to hold a reference of the csv file of data from NLSY79 
file_handle = open(r"C:\Users\15853\Desktop\repos\gorman-earlyjobskill-analysis\data\src\raw-data-from-NLSY79.csv")
```

```python
# Create a file reader for the file handle; DictReader will produce something that is iterable by remembering the first row of the csv is producing the keys of the dictionary  
csv_reader = DictReader(file_handle)
```

```python
# Set up table of data--the type of table being a list of dictionaries with string and float values associated with them.
# the initial table is an empty list  
table: List[Dict[str, float]] = []
```

```python
# Create a table where each row that has float values reads as float values and not strings 
# row is a dictionary from str to str meaning it contains the key which is our column name and the value associated with it for a given row 
for row in csv_reader:
    float_row: Dict[str, float] = {}
    for column in row:
        float_row[column] = float(row[column])
    table.append(float_row)
```
